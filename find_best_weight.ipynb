```python
import os
import json
import pandas as pd
```

```python
weights_json = os.listdir('./runs/test/slerp30_qnorm/')
```

```python
df = pd.DataFrame()
for weight_json in weights_json:
    
    with open(os.path.join('runs/test/slerp30_qnorm', weight_json), 'r') as f:
        log = json.load(f)
        weight_epoch = int(weight_json.split('-')[1].split('.')[0])
        log['epoch'] =  weight_epoch
        df = df.append(log, ignore_index=True)
```
